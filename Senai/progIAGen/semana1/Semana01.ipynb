```python
#Classificação de notas Detalhadas

notaAluno=float(input("Digite a nota do aluno: "))

if notaAluno>=9.0:
  print("Resultado excelente")
elif notaAluno>=7.0:
  print("Bom desempenho")
elif notaAluno>=5.0:
  print("Passou mas foi por pouco")
else:
  print ('Reprovado')
```

```python
# Atividade semana 1

import random

num = random.randint(0, 10)
loop=True
while loop:
  chute_usuario=int(input("Digite um chute para o número sorteado: "))
  if num>chute_usuario:
    print("O número é maior.")
  elif num<chute_usuario:
    print("O número é menor.")
  else:
    print("Parabéns, você acertou.")
    loop=False



```

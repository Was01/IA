```python
# Importação e instalação de bibliotecas

!pip install -q -U google-generativeai

import google.generativeai as genai
from google.colab import userdata
```

```python
CHAVE_API = userdata.get('GOOGLE_API_KEY')

genai.configure(api_key=CHAVE_API)
```

```python
model=genai.GenerativeModel('gemini-2.5-flash-lite')
```

```python
response=model.generate_content("Qual o ano de fundação do senai? Onde está localizado a primeira unidade do Senai? ")

print(response.text)
```

```python

resp = 'continuar'
print("Olá, seja bem-vindo ao chatbot!\n")

while resp != "sair":

    print("Faça sua pergunta: ")
    pergunta = input("")


    try:

        response = model.generate_content(pergunta)


        print(f"\nResposta: {response.text}\n")
        print("-" * 40)

    except Exception as e:
        print(f"Erro ao chamar a API: {e}")


    resp = input("Digite 'sair' para encerrar ou aperte ENTER para continuar: ").strip().lower()

    if  resp.strip().lower() == 'sair':
        break



print("\nPrograma encerrado. Até logo!")
```

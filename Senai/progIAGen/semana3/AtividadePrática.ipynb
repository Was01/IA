```python
# Lendo arquivo csv com pandas

df=pd.read_csv('dataset_atividade_rede_neural_industrial.csv')

print(df.head())
print(df.shape)
```

# Criando uma rede Neuram com Sklearn

```python
# Importando as bibliotecas


from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# Definindo as variáveis

x=df.drop('falha',axis=1)
y=df['falha']
```

```python
# Separando os dados de treinamento e teste

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
# Ajusta o escalonador nos dados de treino e transforma
x_train_scaled = scaler.fit_transform(x_train)
# Apenas transforma os dados de teste (sem ajustar novamente)
x_test_scaled = scaler.transform(x_test)
```

```python
# Criando o modelo

model=MLPClassifier(hidden_layer_sizes=(5,),activation='relu',max_iter=100000,random_state=42)
```

```python
# Treinando o modelo
```

```python
model.fit(x_train_scaled,y_train)
```

```python
# Calculando a acurácia

y_pred=model.predict(x_test_scaled)
accuracy=accuracy_score(y_test,y_pred)
print('Acurácia:',accuracy)
```

# Relatório

1. Contexto do Experimento: O objetivo deste trabalho foi desenvolver e avaliar uma Rede Neural Artificial (Multilayer Perceptron - MLP) para prever falhas em componentes industriais. O conjunto de dados original conta com 150 amostras tendo como variáveis: temperatura, vibração, pressão, corrente do motor, tempo de operação e falha.

2. Configuração da Rede Neural: MLPClassifier (Scikit-Learn)
Arquitetura: 1 camada oculta com 5 neurônios
Função de Ativação: ReLU  
Divisão dos Dados: 70% para treinamento e 30% para testes

3. Resultado: O modelo atingiu uma acurácia de aproximadamente 88,89% nos dados de teste. Em termos práticos, isso significa que a cada 100 diagnósticos que a rede neural faz sobre as máquinas da fábrica, ela acerta cerca de 89 deles (identificando corretamente tanto os casos em que a máquina opera normalmente quanto os casos em que ela apresenta uma falha). Uma acurácia de ~89% indica que as variáveis coletadas possuem forte correlação com o estado de falha, permitindo que o algoritmo aprenda os padrões operacionais com eficiência

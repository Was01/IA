```python
import pandas as pd
```

```python
# Carregando arquivo com pandas

df=pd.read_csv('exercicio.csv')
print(df)
```

```python
# Verificando informações do dataset

print(df.info())
```

```python
# Estatística básica do dataset

print(df.describe())
```

```python
#Identificando valores faltantes
print(df.isnull().sum())
```

```python
# Verificando dados duplicados

print(df.duplicated().sum())
```

```python
# Prenchendo valores faltantes com a média da respectiva coluna


df.fillna(df.mean(), inplace=True)


df_tratado = df.copy()
print(df_tratado)
```

```python
# Excluindo as duplicatas

df_tratado=df.drop_duplicates().copy()
print(df_tratado)
```

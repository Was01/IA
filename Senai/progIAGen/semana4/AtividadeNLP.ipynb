```python
# Lista com 10 palavras

textos=["A matemática é o idioma oculto por trás de cada linha de código. ",
        "Da Inteligência Artificial aos smartphones, tudo nasce de uma equação.",
        "Algoritmos são apenas fórmulas matemáticas resolvendo problemas do mundo real.",
        "A tecnologia avança na velocidade em que deciframos novos padrões numéricos.",
        "Criptografia é a matemática pura garantindo a segurança dos nossos dados digitais.",
        "Os pixels da sua tela são coordenadas geométricas em perfeita harmonia.",
        "Programar é transformar lógica matemática em comandos que o computador entende.",
        "O universo dos games é construído sobre matrizes, vetores e física calculada.",
      "Big Data nada mais é do que estatística matemática aplicada a volumes gigantes de informação.",
       " Sem a matemática, o hardware seria apenas metal e a tecnologia, apenas um conceito."]

for texto in textos:
  print(texto)
```

```python
# Conversão de todas a palavras para minúsculas

textos_minusculos = [texto.lower() for texto in textos]
for texto in textos_minusculos:
  print(texto)
```

```python
# Remoção da acentuação

import re

textos_limpos=[re.sub(r'[^\w\s]', '', texto) for texto in textos_minusculos]

for texto in textos_limpos:
  print(texto)

```

```python
# Tokenizar as palavras

tokens=[texto.split() for texto in textos_limpos]

for token in tokens:
  print(token)
```

```python
from collections import Counter  # Importa a classe Counter, especializada em contar elementos de uma lista


for i, texto in enumerate(tokens):


    contagem = Counter(texto)


    print(f"Contagem de palavras no texto {i+1}:")

    print(contagem)


    print()
```

```python
from sklearn.feature_extraction.text import CountVectorizer  # Importa a ferramenta de vetorização do scikit-learn

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(textos_limpos)


print(vectorizer.get_feature_names_out())


print(X.toarray())
```

```python
import pandas as pd  # Importa a biblioteca Pandas, usada para manipulação e análise de dados

# Cria um DataFrame (uma tabela estruturada) onde:
# - Os dados das células (linhas) vêm da matriz numérica da contagem de palavras (X.toarray())
# - Os nomes das colunas vêm da lista de palavras únicas do vocabulário (vectorizer.get_feature_names_out())
df_bow = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Exibe o DataFrame na tela (em ambientes como o Jupyter Notebook ou Google Colab,
# apenas digitar o nome da variável no final formata os dados como uma tabela visual bonita)
df_bow
```

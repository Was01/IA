```python
textos = [
    "A inteligência artificial está presente na indústria.",
    "A indústria utiliza dados e inteligência artificial.",
    "Modelos de linguagem ajudam na análise de textos."
]

for texto in textos:
  print(texto)
```

```python
texto_minusculo = [texto.lower() for texto in textos]

for texto in texto_minusculo:
  print(texto)
```

## **texto**.lower() = Transforma todas as letras da frase em minúscula.

```python
import re  # Importa a biblioteca de Expressões Regulares (Regular Expressions) do Python

# Cria uma nova lista chamada 'textos_limpos' usando list comprehension.
# Para cada 'texto' dentro da lista 'texto_minusculo', o código usa a função re.sub().
# O padrão r'[^\w\s]' significa: procure por tudo que NÃO seja uma letra, número ([^\w]) ou espaço ([^\s]).
# Tudo o que se encaixar nesse padrão é substituído por uma string vazia '', ou seja, é apagado.

textos_limpos = [re.sub(r'[^\w\s]', '', texto) for texto in texto_minusculo]

# Inicia um laço de repetição (loop) para percorrer cada item da lista de textos já limpos

for texto in textos_limpos:
    print(texto)  # Exibe (imprime) o texto limpo na tela
```

```python
# Cria uma nova lista chamada 'tokens' usando list comprehension.
# Para cada 'texto' dentro de 'textos_limpos', o método .split() divide a frase
# em uma lista de palavras, usando os espaços em branco como separadores.
tokens = [texto.split() for texto in textos_limpos]

# Inicia um loop para percorrer a lista de tokens (que agora é uma lista de listas)
for lista_tokens in tokens:
    print(lista_tokens)  # Exibe cada lista de palavras individualmente na tela
```

```python
from collections import Counter  # Importa a classe Counter, especializada em contar elementos de uma lista

# O 'enumerate' percorre a lista 'tokens', trazendo duas informações a cada volta do loop:
# 'i': o índice (posição) do texto na lista (começando em 0)
# 'texto': a lista de palavras (tokens) daquele texto específico
for i, texto in enumerate(tokens):

    # Cria um objeto Counter que conta a frequência de cada palavra dentro da lista 'texto'
    contagem = Counter(texto)

    # Exibe um cabeçalho identificando qual texto está sendo analisado (somamos +1 para não começar do texto 0)
    print(f"Contagem de palavras no texto {i+1}:")

    # Exibe o resultado da contagem (ex: Counter({'palavra': 3, 'outra': 1}))
    print(contagem)

    # Imprime uma linha em branco apenas para organizar visualmente a saída no terminal
    print()
```

Counter = conta quantas vezes cada palavra aparece

```python
from sklearn.feature_extraction.text import CountVectorizer  # Importa a ferramenta de vetorização do scikit-learn

# Inicializa o vetorizador. É ele quem vai aprender o "vocabulário" dos seus textos.
vectorizer = CountVectorizer()

# O método fit_transform faz duas coisas de uma vez só:
# 1. 'fit': analisa 'textos_limpos' e descobre todas as palavras únicas existentes (o vocabulário).
# 2. 'transform': transforma as frases em uma matriz numérica de contagem de palavras.
X = vectorizer.fit_transform(textos_limpos)

# get_feature_names_out() retorna a lista de todas as palavras que o vetorizador aprendeu.
# Elas funcionam como os "nomes das colunas" da sua matriz e vêm organizadas em ordem alfabética.
print(vectorizer.get_feature_names_out())

# Por padrão, 'X' é uma matriz esparsa (otimizada para economizar memória).
# O método .toarray() transforma 'X' em uma matriz comum (array do NumPy) para que possamos visualizar todas as linhas e colunas de números.
print(X.toarray())
```

CountVectorizer = Transforma os textos em números, binários ou não.

```python
import pandas as pd  # Importa a biblioteca Pandas, usada para manipulação e análise de dados

# Cria um DataFrame (uma tabela estruturada) onde:
# - Os dados das células (linhas) vêm da matriz numérica da contagem de palavras (X.toarray())
# - Os nomes das colunas vêm da lista de palavras únicas do vocabulário (vectorizer.get_feature_names_out())
df_bow = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Exibe o DataFrame na tela (em ambientes como o Jupyter Notebook ou Google Colab,
# apenas digitar o nome da variável no final formata os dados como uma tabela visual bonita)
df_bow
```

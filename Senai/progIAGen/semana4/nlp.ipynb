```python
import pandas as pd
```

```python
textos = [
    "A inteligência artificial está presente na indústria.",
    "A indústria utiliza dados e inteligência artificial.",
    "Modelos de linguagem ajudam na análise de textos."
]
```

```python
df=pd.DataFrame(columns=['Texto'])

df['Texto']=textos

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(df['Texto'])

print(vectorizer.get_feature_names_out())

print(X.toarray())
```

```python
contagem_palavras = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

contagem_palavras
```

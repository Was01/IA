```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import datasets
%matplotlib inline
```

```python
dados=np.array([[7,3,9],[4,6,11],[4,2,5],[5,5,7]])
df=pd.DataFrame(dados,index=['a','b','c','d'],columns=['x','y','z'])
df
```

```python
means_values=df.mean()
means_values
```

```python
var_values=df.var()
var_values
```

```python
cov=df.cov()
cov
```

```python
corr=df.corr()
corr
```

```python
pd.plotting.scatter_matrix(df[['x','y','z']])
plt.show()
```

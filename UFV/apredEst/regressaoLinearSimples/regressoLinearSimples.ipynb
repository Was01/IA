```python
import numpy as np
import pandas as pd
```

```python
popEst=[2,6,8,8,12,16,20,20,22,26]
vendTrim=[58,105,88,118,117,137,157,169,149,202]
```

```python
df=pd.DataFrame({'PopEst':popEst,'VendTrim':vendTrim})
df
```

```python
import seaborn as sns
sns.scatterplot(x='PopEst',y='VendTrim',data=df)
```

```python
import statsmodels.api as sm

x=df['PopEst']
y=df['VendTrim']
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
model.summary()
```

```python
## Predição de um novo valor

novo =pd.DataFrame({'const':[1,1],'PopEst':[10,20]})

model.predict(novo)
```

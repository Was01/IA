```python


df_csv=pd.read_csv('/content/drive/MyDrive/programacao/IA/WRKits/dados/corpus.csv')

df_csv
```

```python
corpus=pd.DataFrame(columns=['Text'])

corpus['Text']=df_csv.iloc[:,0]

corpus
```

```python
# 1. Cria a ferramenta que vai transformar o texto em uma "sacola de palavras" (Bag of Words).
# Por padrão, ele quebra o texto em palavras individuais, remove pontuações e deixa tudo em minúsculo.
vectorizer = CountVectorizer()

# 2. Faz duas coisas ao mesmo tempo:
# 'fit': Analisa a coluna 'Text' do seu banco de dados (corpus) para aprender todas as palavras únicas que existem ali.
# 'transform': Transforma cada texto em uma lista de números, contando quantas vezes cada palavra aprendida aparece.
# O resultado 'X' é salvo como uma matriz esparsa (formato otimizado que economiza memória).
X = vectorizer.fit_transform(corpus['Text'])

# 3. Extrai a lista com os nomes de todas as palavras únicas que o modelo aprendeu no passo anterior.
# Essa lista (tokens) vai estar exatamente na mesma ordem das colunas da matriz 'X'.
tokens = vectorizer.get_feature_names_out()
```
